```python
import pandas as pd
df=pd.read_csv('household_power_consumption.txt',delimiter=';')
```

```python
df['date_time'] = pd.to_datetime(df['Date'].astype(str) + ' ' +df['Time'].astype(str))
```

```python
df=df.dropna()
```

```python
df.reset_index(drop=True)
```

```python
new_df=df.drop(columns=['Voltage','Global_intensity','Sub_metering_1','Sub_metering_2'])
```

```python
(new_df['Global_active_power'][0])*2
```

```python
new_df['Power']=((pd.to_numeric(new_df['Global_active_power'],errors='coerce'))**2+(pd.to_numeric(new_df['Global_reactive_power'],errors='coerce'))**2)**0.5
```

```python
import matplotlib as plt
new_df.plot(x='date_time',y='Power')
```

```python
new_df['month']=new_df['date_time'].dt.month
```

```python
new_df.groupby(['month'])['Power'].mean().plot(kind='bar')
```
